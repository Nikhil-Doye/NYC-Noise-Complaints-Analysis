Load Data

```python
import pandas as pd

# Load the filtered CSV file
df = pd.read_csv('311_Service_Requests_from_2010_to_Present_20240624.csv')
```

```python
df.head()
```

Exploratory Data Analysis

```python
# Display basic information about the dataset
print(df.info())
```

```python
# Display summary statistics for numeric columns
print(df.describe())
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count of complaints by type
complaint_type_counts = df['Complaint Type'].value_counts()
print(complaint_type_counts)

# Visualize the distribution of complaint types
plt.figure(figsize=(20, 30))
sns.countplot(data=df, y='Complaint Type', order=complaint_type_counts.index)
plt.title('Distribution of Noise Complaint Types')
plt.xlabel('Count')
plt.ylabel('Complaint Type')
plt.show()
```

```python
# Count of complaints by borough
borough_counts = df['Borough'].value_counts()
print(borough_counts)

# Visualize the distribution of complaints by borough
plt.figure(figsize=(8, 6))
sns.countplot(data=df, y='Borough', order=borough_counts.index)
plt.title('Distribution of Noise Complaints by Borough')
plt.xlabel('Count')
plt.ylabel('Borough')
plt.show()
```

```python
# Convert 'created_date' to datetime
df['Created Date'] = pd.to_datetime(df['Created Date'], errors='coerce')

# Plot the number of complaints over time
df.set_index('Created Date').resample('M').size().plot(title='Monthly Noise Complaints', figsize=(12, 6))
plt.xlabel('Date')
plt.ylabel('Number of Complaints')
plt.show()
```

```python
# Extract hour and day of the week from 'created_date'
df['hour'] = df['Created Date'].dt.hour
df['day_of_week'] = df['Created Date'].dt.day_name()

# Visualize complaints by hour of the day
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='hour')
plt.title('Noise Complaints by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Count')
plt.show()

# Visualize complaints by day of the week
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='day_of_week', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Noise Complaints by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Count')
plt.show()
```

Data Cleaning

```python
# Drop rows with missing values in critical columns
df_cleaned = df.dropna(subset=['Created Date', 'Complaint Type', 'Incident Zip', 'Borough', 'Latitude', 'Longitude'])

# Convert 'created_date' and 'closed_date' to datetime
df_cleaned['Closed Date'] = pd.to_datetime(df_cleaned['Closed Date'])

# Display data types to ensure correctness
print(df_cleaned.dtypes)

# Save cleaned data for further analysis
df_cleaned.to_csv('cleaned_nyc_noise_complaints.csv', index=False)
```

Data Visualization

```python
plt.figure(figsize=(12, 25))
sns.countplot(data=df_cleaned, y='Complaint Type', order=complaint_type_counts.index)
plt.title('Number of Noise Complaints by Type')
plt.xlabel('Count')
plt.ylabel('Complaint Type')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df_cleaned.set_index('Created Date').resample('M').size().plot()
plt.title('Trend of Noise Complaints Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Complaints')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df_cleaned['Borough'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Noise Complaints by Borough')
plt.ylabel('')
plt.show()
```

```python
heatmap_data = df_cleaned.groupby(['day_of_week', 'hour']).size().unstack().fillna(0)

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='YlGnBu')
plt.title('Heatmap of Noise Complaints by Hour and Day of Week')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plt.show()
```

```python
import folium
from folium.plugins import HeatMap

# Create a base map
m = folium.Map(location=[40.7128, -74.0060], zoom_start=11)

# Prepare data for heatmap
heat_data = [[row['Latitude'], row['Longitude']] for index, row in df_cleaned.iterrows() if not pd.isnull(row['Latitude'])]

# Add heatmap to the map
HeatMap(heat_data).add_to(m)

# Save map to HTML
m.save('nyc_noise_complaints_heatmap.html')
```
